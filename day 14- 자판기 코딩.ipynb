```python
money = 0
pennies = 0.01
dime = 0.1
nickel = 0.05
quater = 0.25
```

```python
#espresoo
def espresso():
    
    print("돈 넣어줘")
    
    global money
    global pennies
    global dime
    global nickel
    global quater
    
    cost = 0
    quat =0
    nic =0
    dim =0
    pen = 0
    
    
    quat += int(input("how many quaters : "))
    nic += int(input("how many nic : "))
    dim += int(input("how many dim : "))
    pen += int(input("how many pen : "))
    
    while(cost < MENU['espresso']['cost']):
        cost = quat *quater + nic *nickel + dim *dime + pen *pennies
        if cost >=MENU['espresso']['cost']:
            if MENU['espresso']['ingredients']['water'] > resources['water']:
                print("물 부족")
            elif MENU['espresso']['ingredients']['coffee'] > resources['coffee'] :
                print("커피 부족")
            else: 
                print(f"here is : ${round(cost - MENU['espresso']['cost'],2)}")
                print("여기 에스프레소 나왔습니다")
                money +=  MENU['espresso']['cost']
        else:
            print("돈 더주세요")
            quat += int(input("how many quaters"))
            nic += int(input("how many nic"))
            dim += int(input("how many dim"))
            pen += int(input("how many pen"))
```

```python
def latte():
    
    print("돈 넣어줘")
    
    global money
    global pennies
    global dime
    global nickel
    global quater
    
    cost = 0
    quat =0
    nic =0
    dim =0
    pen = 0
    
    
    quat += int(input("how many quaters : "))
    nic += int(input("how many nic : "))
    dim += int(input("how many dim : "))
    pen += int(input("how many pen : "))
    
    while(cost < MENU['latte']['cost']):
        cost = quat *quater + nic *nickel + dim *dime + pen *pennies
        if cost >=MENU['latte']['cost']:
            if MENU['latte']['ingredients']['water'] > resources['water']:
                print("물 부족")
            elif MENU['latte']['ingredients']['milk']  > resources['milk']:
                print("우유 부족")
            elif MENU['latte']['ingredients']['coffee'] > resources['coffee'] :
                print("커피 부족")
            
            else: 
                print(f"here is : ${round(cost - MENU['latte']['cost'],2)}")
                print("여기 라때 나왔습니다")
                money +=  MENU['latte']['cost']
        else:
            print("돈 더주세요")
            quat += int(input("how many quaters"))
            nic += int(input("how many nic"))
            dim += int(input("how many dim"))
            pen += int(input("how many pen"))
            
```

```python
def cappuccino():
    
    print("돈 넣어줘")
    
    global money
    global pennies
    global dime
    global nickel
    global quater
    
    cost = 0
    quat =0
    nic =0
    dim =0
    pen = 0
    
    
    quat += int(input("how many quaters : "))
    nic += int(input("how many nic : "))
    dim += int(input("how many dim : "))
    pen += int(input("how many pen : "))
    
    while(cost < MENU['cappuccino']['cost']):
        cost = quat *quater + nic *nickel + dim *dime + pen *pennies
        if cost >=MENU['cappuccino']['cost']:
            if MENU['cappuccino']['ingredients']['water'] > resources['water']:
                print("물 부족")
            elif MENU['cappuccino']['ingredients']['milk']  > resources['milk']:
                print("우유 부족")
            elif MENU['cappuccino']['ingredients']['coffee'] > resources['coffee'] :
                print("커피 부족")
            
            else: 
                print(f"here is : ${round(cost - MENU['cappuccino']['cost'],2)}")
                print("여기 카푸치노 나왔습니다")
                money +=  MENU['cappuccino']['cost']
        else:
            print("돈 더주세요")
            quat += int(input("how many quaters"))
            nic += int(input("how many nic"))
            dim += int(input("how many dim"))
            pen += int(input("how many pen"))
            
    
    
```

```python
# 남은 재료 보기
def report():
    global money 
    print(f"water : {resources['water']}ml")
    print(f"milk : {resources['milk']}ml")
    print(f"coffee : {resources['coffee']}g")
    print(f"money : ${money}")
    
```

```python
# r
def report_minut():
    global drink
    if drink == 'espresso':
        resources['water'] -= MENU['espresso']['ingredients']['water'] 
        resources['coffee'] -= MENU['espresso']['ingredients']['coffee']
    elif drink == 'latte':
        resources['water'] -= MENU['latte']['ingredients']['water'] 
        resources['milk'] -= MENU['latte']['ingredients']['coffee']
        resources['coffee'] -= MENU['latte']['ingredients']['coffee']
    else:
        resources['water'] -= MENU['cappuccino']['ingredients']['water'] 
        resources['milk'] -= MENU['cappuccino']['ingredients']['coffee']
        resources['coffee'] -= MENU['cappuccino']['ingredients']['coffee']      
```

```python
# 메인
drink = input("espresso or latte or cappuccino : ")
if drink == 'report':
    report()
elif drink == 'espresso':
    espresso()
    report_minut()
elif drink =='latte':
    latte()
    report_minut()
else:
    cappuccino()
    report_minut()
```
